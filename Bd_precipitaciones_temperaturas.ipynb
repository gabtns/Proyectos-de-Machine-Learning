```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv("/content/ny_precipitaciones.csv")
df1=pd.read_csv("/content/ny_temperaturas.csv")
```

```python
df.head()
```

```python
datos2=pd.merge(df,df1,how="inner",on='STATION')#Union de los dataframes
datos2.fillna(0).head()#Reemplazo de los valores nulos por ceros
datos2.set_index("NAME_y")#Nueva indexacion
```

```python
datos2=datos2.drop(labels=["NAME_y","STATION"],axis=1)#Eliminacion de las columnas 
```

```python
datos2.head(3)#Verifiacion de las eliminaciones
```

```python
#Unacolumna basada en la suma de las precipitaciones
datos2['suma_prec_total']=datos2['MLY-PRCP-25PCTL']+datos2['MLY-PRCP-50PCTL']+datos2['MLY-PRCP-75PCTL']
```

```python
#Al tener las temperaturas en fahrenheit, se procede hacer la funcion para convertir los datos a celsius
def cambiar(x):
  return (x - 32)*0.5556 
```

```python
#A las columnas se procede apliacar dicha funcion, dandole solamente 2 decimales para acortar los resultados
datos2['tem_normal']=round(datos2['MLY-TAVG-NORMAL'].apply(cambiar),2)
datos2['tem_max_normal']=round(datos2['MLY-TMAX-NORMAL'].apply(cambiar),2)
datos2['tem_min']=round(datos2['MLY-TMIN-NORMAL'].apply(cambiar),2)
```

```python
#Eliminacion de las viejas columnas
datos2=datos2.drop(labels=["MLY-TAVG-NORMAL","MLY-TMAX-NORMAL","MLY-TMIN-NORMAL"], axis=1, index=None, columns=None, level=None, inplace=False)
```

```python
datos2=datos2.dropna(axis=0)#Eliminacion de los datos faltantes
```

```python
fig=plt.figure(figsize=(15,7))
sns.relplot(x='DATE_x',y="suma_prec_total",data=datos2,hue="MLY-PRCP-50PCTL")
plt.show()
```

```python
#Agrupamiento de los datos de temperaturas indexados por los meses
temperaturas=datos2.groupby(by="DATE_y")[["tem_normal","tem_max_normal","tem_min"]].mean()
temperaturas
```

```python
#Precipitaciones en 3D,clasificadas por los parametros normales
px.scatter_3d(datos2,x="DATE_y",y="MLY-TMAX-STDDEV",z="tem_normal",
              labels={"DATE_y":"Mes","MLY-TMAX-STDDEV":"Valores normales","tem_normal":"temperaturas normales"},color="MLY-TMAX-STDDEV")
```

```python
#Mapa de calor de las temperaturas registradas
fig=plt.figure(figsize=(15,7))
sns.heatmap(temperaturas,xticklabels="auto", yticklabels="auto",)
```

```python
from math import degrees
from matplotlib.cbook import get_label
fig=plt.figure(figsize=(25,10))
fig,(xa1,xa2,xa3)=plt.subplots(3,1,tight_layout=False)
fig=fig.add_axes([0,0,1,1])
x=datos2['DATE_x']
y=datos2['MLY-PRCP-25PCTL']

figura_0=fig.plot(x,y,color="red",label="25")

x1=datos2['DATE_x']
x2=datos2['MLY-PRCP-50PCTL']
figura=fig.plot(x1,x2,color="orange",label="50")

j=datos2['DATE_x']
j1=datos2['MLY-PRCP-75PCTL']
figura_1=fig.plot(j,j1,color="yellow",label="75")

plt.title("Precipitaciones")
plt.xlabel("Meses")
plt.ylabel("Volumen de precipitaciones")
funciones=figura+figura_0+figura_1
labels=[i.get_label() for i in funciones]
  
plt.legend(funciones,labels,loc="upper right")
```

```python
precio=datos2.groupby(by="DATE_x")["MLY-PRCP-25PCTL","MLY-PRCP-50PCTL","MLY-PRCP-75PCTL"].sum()
```

```python
fig=plt.figure(figsize=(10,8))

sns.heatmap(precio,cmap="YlGnBu",cbar=True,annot=False)
```
